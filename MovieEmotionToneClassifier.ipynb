```python
# Import necessary libraries
import pandas as pd
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
!pip install gradio
import gradio as gr
```

```python
# Load dataset
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/imdb_movies.csv')
```

```python
df.head()
```

```python
# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
def classify_movie_emotion(movie_title):
    # Filter dataset based on the input movie title
    movie_data = df[df['orig_title'].str.lower() == movie_title.lower()]

    if movie_data.empty:
        return "No movie found with the specified title."

    # Extract movie overview
    movie_overview = movie_data.iloc[0]['overview']

    # Perform sentiment analysis on the movie overview
    sentiment_score = analyzer.polarity_scores(movie_overview)['compound']

    # Classify emotional tone based on sentiment score
    if sentiment_score >= 0.2:
        emotional_tone = "Happiness"
    elif sentiment_score >= 0 and sentiment_score < 0.2:
        emotional_tone = "Laughter"
    elif sentiment_score >= -0.2 and sentiment_score < 0:
        emotional_tone = "Neutral"
    elif sentiment_score >= -0.5 and sentiment_score < -0.2:
        emotional_tone = "Sadness"
    else:
        emotional_tone = "Fear"

    # Output emotional tone score
    emotional_tone_score = {
        "Movie Title": movie_data.iloc[0]['orig_title'],
        "Emotional Tone": emotional_tone,
        "Sentiment Score": sentiment_score
    }

    return emotional_tone_score
```

```python
# Create Gradio interface
input_movie_title = gr.Textbox(label="Enter a movie title:")
output_emotion = gr.Textbox(label="Movie Emotion Rating")

gr.Interface(fn=classify_movie_emotion, inputs=input_movie_title, outputs=output_emotion,
             title="Movie Emotion Rating",
             description="Enter a movie title to classify the emotional tone and sentiment of the movie's overview.").launch()
```
